```python
import gensim
import numpy as np
import pandas as pd
```

```python
ftt_model = gensim.models.Word2Vec.load("FastTest.kv")
```

```python
x = ftt_model.wv.vectors
print(x)
```

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from sklearn.decomposition import PCA
```

```python
X_ftt = list(ftt_model.wv[ftt_model.wv.key_to_index.keys()])
ftt_words = list(ftt_model.wv.key_to_index.keys())
```

```python
pca_ftt = PCA(n_components=2)
result = pca_ftt.fit_transform(X_ftt)

#create df from the pca results
pca_df_ftt = pd.DataFrame(result, columns = ['x','y'])

#add the words for the hover effect
pca_df_ftt['word'] = ftt_words
pca_df_ftt.head()
```

```python
fig = go.Figure(data=go.Scattergl(
   x = pca_df_ftt['x'],
   y = pca_df_ftt['y'],
   mode='markers',
   marker=dict(
       color=np.random.randn(100000),
       colorscale='rainbow',
       line_width=1
   ),
   text=pca_df_ftt['word'],
   textposition="bottom center",
))

fig.show()
```
